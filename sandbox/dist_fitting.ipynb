```python
import pyro.distributions as dist
import torch
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Fit normal to uniform-sampled data. What's it look like?
data_uniform = dist.Uniform(torch.ones(1)*0.4, torch.ones(1)*0.6).sample(sample_shape=(10,))
data_normal_fit = dist.Normal(
    torch.mean(data_uniform),
    torch.sqrt(torch.var(data_uniform))
).sample(sample_shape=(1000,))
plt.figure()
plt.hist(data_uniform.flatten().numpy(), label="Uniform samples", alpha=0.5, density=True)
plt.hist(data_normal_fit.flatten().numpy(), label="Samples from normal fit to uniform samples", alpha=0.5, density=True)
plt.legend()
```
